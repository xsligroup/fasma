```python
import os
data_dir = os.path.abspath('').replace("notebooks", "data\\")
```

```python
box = fc.parse(os.path.join(data_dir, "water_td-rhf.log"))
```

We should now have the relevant data from this population calculation.

We make a call to the generate_mo_analysis function, which returns a pandas Dataframe if this file contains a population calculation.

```python
data = box.generate_mo_analysis()
```

```python
data
```

Imagine if we have a hundred or more MOs. Our dataframe would be unnecessarily large, especially when people want to care about a few specific MOs that have contribution for a characteristic. We can call show_nonzero_mo on any dataframe to assist our analysis as it lists all the MOs with non-zero electron contribution.

```python
dff.show_nonzero_mo(data)
```

We can also filter out attributes that we don't care about by calling filter_mo_analysis() which takes a dataframe, a list of attributes (index), and a list of MOs (mo_list) to keep in the new filtered dataframe. If an empty list is provided for either the index list or mo list, all the columns associated with that list will be returned.

```python
dff.filter_mo_analysis(data, ["Atom Type", "Subshell"])
```

Here, we group by atom type, principal quantum number, and atomic orbital.

```python
dff.filter_mo_analysis(data, ["Atom Type", "Principal Quantum Number", "Atomic Orbital"])
```

Let's look exclusively at the first three MOs.

```python
dff.filter_mo_analysis(data, mo_list=[1, 2, 3])
```

By default, generate_mo_analysis() creates its dataframe from looking at the alpha electron MO Coefficients. Should your calculation be a UHF calculation, you can do an MO analysis using the beta MO Coefficients as shown below.

```python
box2 = fc.parse(os.path.join(data_dir, "Na_uhf.log"))
```

```python
data2 = box2.generate_mo_analysis(electron='beta')
```

```python
data2
```

Let's compare that to the normal MO analysis (by default, alpha is returned).

```python
data3 = box.generate_mo_analysis()
```

```python
data3
```

The examples shown here are not exhaustive of the possible MO analyses.
